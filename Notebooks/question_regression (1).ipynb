```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the house_price_data table

# 1. Average price of houses grouped by bedrooms
average_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().reset_index()
average_price_by_bedrooms.columns = ['bedrooms', 'AveragePrice']

# 2. Average sqft_living of houses grouped by bedrooms
average_sqft_living_by_bedrooms = df.groupby('bedrooms')['sqft_living'].mean().reset_index()
average_sqft_living_by_bedrooms.columns = ['bedrooms', 'AverageSqftLiving']

# 3. Average price of houses with and without waterfront
average_price_by_waterfront = df.groupby('waterfront')['price'].mean().reset_index()
average_price_by_waterfront.columns = ['waterfront', 'AveragePrice']

# 4. Correlation between the columns condition and grade
correlation_condition_grade = df.groupby('condition')['grade'].agg('corr', df['grade']).reset_index()
correlation_condition_grade.columns = ['condition', 'Correlation']

# Display the results
print("1. Average price of houses grouped by bedrooms:")
print(average_price_by_bedrooms)

print("\n2. Average sqft_living of houses grouped by bedrooms:")
print(average_sqft_living_by_bedrooms)

print("\n3. Average price of houses with and without waterfront:")
print(average_price_by_waterfront)

print("\n4. Correlation between the columns condition and grade:")
print(correlation_condition_grade)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the house_price_data table

# Convert 'floors' and 'bathrooms' columns to integer
df['floors'] = df['floors'].astype(int)
df['bathrooms'] = df['bathrooms'].astype(int)

# Filter the DataFrame based on customer criteria
filtered_options = df[
    (df['bedrooms'].isin([3, 4])) &
    (df['bathrooms'] > 3) &
    (df['floors'] == 1) &
    (df['waterfront'] == 0) &
    (df['condition'] >= 3) &
    (df['grade'] >= 5) &
    (df['price'] < 300000)
]

# Display the result
print("Options available for the customer:")
print(filtered_options)

```

```python
import pandas as pd


# 1. Calculate the average price of all properties
average_price = df['price'].mean()

# 2. Filter properties whose prices are twice more than the average
twice_avg_price_properties = df[df['price'] > 2 * average_price]

# Display the result
print("List of properties whose prices are twice more than the average:")
twice_avg_price_properties
```

```python
def twice_avg_price_view(dataframe):
    # Calculate the average price of all properties
    average_price = dataframe['price'].mean()

    # Filter properties whose prices are twice more than the average
    twice_avg_price_properties = dataframe[dataframe['price'] > 2 * average_price]

    return twice_avg_price_properties

# Create a view by applying the function to the DataFrame
twice_avg_price_view_df = twice_avg_price_view(df)

# Display the view
print("View of properties whose prices are twice more than the average:")

twice_avg_price_view_df
```

```python
# Calculate the average prices for properties with three and four bedrooms
average_price_3_bedrooms = df[df['bedrooms'] == 3]['price'].mean()
average_price_4_bedrooms = df[df['bedrooms'] == 4]['price'].mean()

# Calculate the difference in average prices
price_difference = average_price_4_bedrooms - average_price_3_bedrooms

# Display the result
print("Difference in average prices between properties with three and four bedrooms:")
print(price_difference)
```

```python
# Get the distinct zip codes
distinct_zip_codes = df['zipcode'].unique()

# Display the result
print("Different locations (distinct zip codes) where properties are available:")
print(distinct_zip_codes)
```
