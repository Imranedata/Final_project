Is there a significant difference in the average selling prices of houses with waterfront views compared to those without?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
from scipy import stats

# Split the dataset into houses with and without waterfront views
waterfront_houses = df[df['waterfront'] == 1]['price']
non_waterfront_houses = df[df['waterfront'] == 0]['price']

# Perform t-test
t_statistic, p_value = stats.ttest_ind(waterfront_houses, non_waterfront_houses)

# Print results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the average selling prices of houses with and without waterfront views.")
else:
    print("There is no significant difference in the average selling prices of houses with and without waterfront views.")
```

```python
# Plotting
plt.figure(figsize=(8, 6))
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Comparison of Average Selling Prices (Waterfront vs Non-Waterfront)')
plt.xlabel('Waterfront (0: No, 1: Yes)')
plt.ylabel('Price ($)')
plt.xticks([0, 1], ['Non-Waterfront', 'Waterfront'])
plt.show()
```

```python
# Filter the DataFrame for houses with and without waterfront views
waterfront_prices = df[df['waterfront'] == 1]['price']
non_waterfront_prices = df[df['waterfront'] == 0]['price']

# Combine the filtered data into a new DataFrame
price_comparison_data = pd.DataFrame({'Waterfront': waterfront_prices, 'Non-Waterfront': non_waterfront_prices})

# Save the data to a CSV file
price_comparison_data.to_csv('price_comparison_waterfront_vs_non_waterfront.csv', index=False)
```
