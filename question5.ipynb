What is the average price per square foot of living space for houses in the dataset, and how does this metric compare across different conditions?

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame containing the housing data

# Calculate price per square foot
df['price_per_sqft'] = df['price'] / df['sqft_living']

# Group by house condition and compute average price per square foot
avg_price_per_sqft_by_condition = df.groupby('condition')['price_per_sqft'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_per_sqft_by_condition, x='condition', y='price_per_sqft', palette='viridis')
plt.title('Average Price per Square Foot by House Condition')
plt.xlabel('House Condition')
plt.ylabel('Average Price per Square Foot')
plt.show()
```
